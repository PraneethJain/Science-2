```python
import seaborn as sns
import matplotlib.pyplot as plt
from random import choice


roll_number = 2022101093
a = -(roll_number % 5 + 1)
b = 2 * (roll_number % 10)
big = 10000
```

```python
def walk_once(start: int, n: int) -> int:
    return start + sum(choice([-1, 1]) for _ in range(n))
```

```python
# (i)
sns.scatterplot([sum(walk_once(a, n) == 0 for _ in range(big))/big for n in range(1, 101)])
plt.show()
```

```python
# (ii)
sns.scatterplot(
    [
        sum(walk_once(a, n) == walk_once(b, n) for _ in range(big)) / big
        for n in range(101, 1001, 50)
    ]
)
plt.show()
```

```python
sns.lineplot([sum(walk_once(0, n) for _ in range(big)) / big for n in range(1, 101)])
plt.show()
```
