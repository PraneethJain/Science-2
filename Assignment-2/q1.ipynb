```python
from random import uniform
from math import sin
import seaborn as sns
import matplotlib.pyplot as plt


def monte_carlo(f, a: float, b: float, N: int) -> float:
    return sum(f(uniform(a, b)) * (b - a) for _ in range(N)) / N
```

## 1.4 (a)

$$ \int_{-1}^{1} (x^2 + 2x)dx $$
$$ [x^3/3 + x^2]_{-1}^{1} $$
$$ (1/3 + 1) - (-1/3 + 1) $$
$$ 2/3 $$


```python
sns.lineplot(x=range(1,5),y=[monte_carlo(lambda x: x*x + 2*x, -1, 1, 10**i) for i in range(1,5)])
plt.show()
```

## 1.4 (b)

$$ \int_{-\pi/2}^{\pi/2} x^2 sin(x) dx $$
$$  \int_{-\pi/2}^{0} x^2 sin(x) dx + \int_{0}^{\pi/2} x^2 sin(x) dx $$
$$  \int_{\pi/2}^{0} (-x)^2 sin(-x) (-dx) + \int_{0}^{\pi/2} x^2 sin(x) dx $$
$$  -\int_{0}^{\pi/2} x^2 sin(x) dx + \int_{0}^{\pi/2} x^2 sin(x) dx $$
$$  0 $$


```python
sns.lineplot(
    x=range(1, 5),
    y=[monte_carlo(lambda x: x * x * sin(x), -1, 1, 10**i) for i in range(1, 5)],
)
plt.show()
```
